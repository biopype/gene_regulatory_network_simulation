# Construction and Analysis of BRCA1 Gene Regulatory Network to study Protein-Protein Interactions and Regulatory Dynamics

# Load the File

The file contains the interaction of BRCA1 genes with other proteins in *Homo sapiens* and corresponding confidence score in addition to other interaction metrics.

```python
import pandas as pd
```

```python
import pandas as pd

tsv_file = "string_interaction.tsv"
csv_file = "string_interaction.csv"

# Read the TSV file into a DataFrame
df = pd.read_csv(tsv_file, sep='\t')

# Save the DataFrame as a CSV file
df.to_csv(csv_file, index=False)

print(f"TSV file has been converted to CSV and saved as {csv_file}")
```

# Network Construction

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add edges with weights (combined_score)
for _, row in df.iterrows():
    G.add_edge(row['#node1'], row['node2'], weight=row['combined_score'])

# Display network properties
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")

# Visualize the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Position nodes using the spring layout
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=1, edge_color='gray', alpha=0.7)
nx.draw_networkx_labels(G, pos, font_size=8)

# Draw edge weights as labels (optional)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7)

plt.title("Weighted Protein Interaction Network")
plt.show()
```

# Inferring Regulatory Interaction Types

**Inferring types (activation or repression) based on these columns:**
* High coexpression: Likely activation.
* High neighborhood_on_chromosome (proximity): Potential co-regulation.
* High gene_fusion or homology: Suggest redundancy.




```python
import pandas as pd

# thresholds for interaction types
high_threshold = {
    'coexpression': 0.06,
    'neighborhood_on_chromosome': 0,
    'gene_fusion': 0,
    'homology': 0.05
}

# Initialize interaction type column
df['interaction_type'] = 'unknown'

# Infer interaction types
for index, row in df.iterrows():
    if row['coexpression'] >= high_threshold['coexpression']:
        df.at[index, 'interaction_type'] = 'activation'
    elif row['neighborhood_on_chromosome'] >= high_threshold['neighborhood_on_chromosome']:
        df.at[index, 'interaction_type'] = 'co-regulation'
    elif row['gene_fusion'] >= high_threshold['gene_fusion'] or row['homology'] >= high_threshold['homology']:
        df.at[index, 'interaction_type'] = 'redundancy'

df.to_csv("interactions_with_types.csv", index=False)
```

# Thresholding and Filtering

```python
import pandas as pd

# thresholds for filtering
combined_score_threshold = 0.8
experiment_weight = 1  # Define weight for experimentally determined interactions
database_weight = 1    # Define weight for database-annotated interactions

# Filter interactions based on combined_score
df_filtered = df[df['combined_score'] >= combined_score_threshold]

# Prioritize experimentally determined and database-annotated interactions
df_filtered['high_confidence'] = (
    (df_filtered['experimentally_determined_interaction'] * experiment_weight) +
    (df_filtered['database_annotated'] * database_weight)
)

# Sort by high-confidence score (descending)
df_prioritized = df_filtered.sort_values(by='high_confidence', ascending=False)

# Save the filtered and prioritized data
output_file = "filtered_prioritized_interactions.csv"
df_prioritized.to_csv(output_file, index=False)
```

# Dynamic Modeling using Ordinary Differential Equations (ODEs)

Each node (protein) in the geneset is considered as a variable, and confidence score measures the strength of interaction. The interaction type (activation or repression) determines whether the relationship is positive or negative.

**Mapping Nodes to Indices:**

The nodes (proteins) are mapped to indices for easy matrix manipulation.

**Interaction Matrix:**

A matrix represents the network:

Rows: Regulators (node1).

Columns: Targets (node2).



**ODE System:**

image.png


```python
import pandas as pd
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

df = df_reset

# Define parameters for modeling
interaction_threshold = 0.7  # Use interactions above this combined_score
time_span = (0, 10)  # Simulation time span
time_points = np.linspace(time_span[0], time_span[1], 500)  # Time points for solution

# Filter high-confidence interactions
df_filtered = df[df['combined_score'] >= interaction_threshold]

# Parse nodes and create a list of unique proteins
nodes = set(df_filtered['#node1']).union(set(df_filtered['node2']))
nodes = list(nodes)

# Map nodes to indices for easier modeling
node_index = {node: i for i, node in enumerate(nodes)}

# Initialize interaction matrix
n = len(nodes)
interaction_matrix = np.zeros((n, n))  # Rows = regulators, Columns = targets

# Populate interaction matrix
for _, row in df_filtered.iterrows():
    source = node_index[row['#node1']]
    target = node_index[row['node2']]
    strength = row['combined_score']

    # Determine activation (+) or repression (-)
    if row['coexpression'] >= 0.7:
        interaction_matrix[source, target] = strength  # Activation
    elif row['gene_fusion'] >= 0.6 or row['homology'] >= 0.5:
        interaction_matrix[source, target] = -strength  # Repression

# Define the system of ODEs
def protein_network(t, concentrations):
    d_concentrations_dt = np.zeros(n)
    for i in range(n):  # For each protein
        regulation = sum(interaction_matrix[j, i] * concentrations[j] for j in range(n))
        # Assume a basic degradation term (-concentrations[i])
        d_concentrations_dt[i] = regulation - concentrations[i]
    return d_concentrations_dt

# Initial concentrations (arbitrary, can adjust based on biology)
initial_concentrations = np.random.rand(n)

# Solve the ODE system
solution = solve_ivp(protein_network, time_span, initial_concentrations, t_eval=time_points)

# Plot results
plt.figure(figsize=(10, 6))
for i, protein in enumerate(nodes):
    plt.plot(solution.t, solution.y[i], label=protein)
plt.xlabel("Time")
plt.ylabel("Protein Concentration")
plt.title("Dynamic Protein Network Model")
plt.legend()
plt.show()
```

# Visualization

Visualization for combined_score

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph using NetworkX
G = nx.Graph()

for _, row in df.iterrows():
    G.add_edge(
        row['#node1'],
        row['node2'],
        weight=row['combined_score']
    )

# Compute node degrees (number of interactions)
node_degrees = dict(G.degree())

# Get node sizes based on degree (scale it for visualization)
node_sizes = [50 + 20 * node_degrees[node] for node in G.nodes]

# Get edge widths based on combined_score
edge_widths = [G[u][v]['weight'] * 10 for u, v in G.edges]

# Draw the graph
plt.figure(figsize=(12, 8))

# Layout for the graph
pos = nx.spring_layout(G, seed=42)  # Spring layout for aesthetic positioning

# Draw nodes
nx.draw_networkx_nodes(
    G,
    pos,
    node_size=node_sizes,
    node_color='skyblue',
    alpha=0.7
)

# Draw edges
nx.draw_networkx_edges(
    G,
    pos,
    width=edge_widths,
    edge_color='gray',
    alpha=0.7
)

# Draw labels
nx.draw_networkx_labels(
    G,
    pos,
    font_size=10,
    font_color='black'
)

# Add a title
plt.title("Protein-Protein Interaction Network", fontsize=16)
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```

Visualization for combined_score, coexpression, gene_fusion

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph using NetworkX
G = nx.Graph()

# Add edges with attributes (combined_score, coexpression, gene_fusion)
for _, row in df.iterrows():
    G.add_edge(
        row['#node1'],
        row['node2'],
        weight=row['combined_score'],  # Edge weight
        coexpression=row['coexpression'],  # Additional metric
        gene_fusion=row['gene_fusion']  # Additional metric
    )

# Compute node degrees (number of interactions)
node_degrees = dict(G.degree())

# Get node sizes based on degree (scale for visualization)
node_sizes = [50 + 20 * node_degrees[node] for node in G.nodes]

# Get edge thickness based on combined_score
edge_widths = [G[u][v]['weight'] * 10 for u, v in G.edges]

# Get edge colors based on coexpression (normalized for visualization)
coexpression_values = [G[u][v]['coexpression'] for u, v in G.edges]
max_coexpression = max(coexpression_values)
edge_colors = [val / max_coexpression for val in coexpression_values]

# Draw the graph
fig, ax = plt.subplots(figsize=(14, 10))  # Create a figure and axis

# Layout for the graph
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw_networkx_nodes(
    G,
    pos,
    ax=ax,  # Use the axis created
    node_size=node_sizes,
    node_color='skyblue',
    alpha=0.8
)

# Draw edges with color and thickness
nx.draw_networkx_edges(
    G,
    pos,
    ax=ax,  # Use the axis created
    width=edge_widths,
    edge_color=edge_colors,
    edge_cmap=plt.cm.plasma,  # Color map for coexpression
    alpha=0.8
)

# Draw labels
nx.draw_networkx_labels(
    G,
    pos,
    ax=ax,
    font_size=10,
    font_color='black'
)

# color bar
sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma, norm=plt.Normalize(vmin=min(coexpression_values), vmax=max_coexpression))
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label("Coexpression Levels", fontsize=12)

ax.set_title("Enhanced Protein-Protein Interaction Network", fontsize=16)
plt.axis('off')
plt.show()
```
